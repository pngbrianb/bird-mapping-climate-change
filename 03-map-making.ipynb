```python
import geopandas as gpd
import pandas as pd
import numpy as np
import os
import fiona
from matplotlib import pyplot as plt
from shapely.geometry import polygon
%matplotlib inline
```

# Final Mapping Notebook

The purpose of this notebook is to read in my grids and map them as chloropleths for the purpose of comparison.

```python
# reading in my aggregate maps in a dictionary in case I want to iterate through them.
raster_dic = {
    'sum_00' : gpd.read_file('./birds-data/summer_00_agg/'),
    'win_00' : gpd.read_file('./birds-data/winter_00_agg/'),
    'sum_20' : gpd.read_file('./birds-data/summer_20_agg/'),
    'win_20' : gpd.read_file('./birds-data/winter_20_agg/'),
    'sum_50' : gpd.read_file('./birds-data/summer_50_agg/'),
    'win_50' : gpd.read_file('./birds-data/winter_50_agg/'),
    'sum_80' : gpd.read_file('./birds-data/summer_80_agg/'),
    'win_80' : gpd.read_file('./birds-data/winter_80_agg/')
}
```

```python
# checking that the coordinates got read in properly
raster_dic['sum_00'].crs
```

```python
raster_dic['sum_00'].head()
```

```python
# reading in my base map and setting its coordinates to match my aggregate maps
northam = gpd.read_file('./NorthAm.json')
northam = northam.to_crs(raster_dic['win_50'].crs)
northam.crs
```

## Mapping and Exporting

Here I simply map each study and export the result as a .png image. I put the summer and winter bird species side-by-side to more easily compare across them all in my presentation.

```python
f, (ax1, ax2) = plt.subplots(1, 2,figsize=(20,9))

raster_dic['sum_00'].plot(ax=ax1,cmap='Reds',column='count',legend=True)
northam.plot(ax=ax1,color='white', edgecolor='black',alpha=0.2);
ax1.set_title('Summer 2000',fontsize=20)
ax1.set_xticklabels([])
ax1.set_yticklabels([])

raster_dic['win_00'].plot(ax=ax2,cmap='Blues',column='count',legend=True)
northam.plot(ax=ax2,color='white', edgecolor='black',alpha=0.2);
ax2.set_title('Winter 2000',fontsize=20)
ax2.set_xticklabels([])
ax2.set_yticklabels([])
plt.savefig('./assets/both_2000.png', bbox_inches='tight')
plt.show();
```
